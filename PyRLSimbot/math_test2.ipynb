```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
coll1 = np.loadtxt("Plot/collision_count.csv")
eat1 = np.loadtxt("Plot/eat_count.csv")

plt.plot(coll1)
```

```python
x = eat1[100:120]
num = 10000000000000
import numpy as np
data = np.round([max(x)*num,min(x)*num] )
print(data)
```

```python
import random

y1=[]
for i in range(300-130):
    y1.append(random.randint(data[1], data[0])/num)
    
# plt.plot(y1)
```

```python
random.randint(10000,11000)
```

```python
y2_gain = []
for i in range(len(y1)):
    y2_gain.append(random.randint(10000,10200)/10000)
```

```python
eat2 = np.append(eat1[:130],np.array(y1)*y2_gain)
```

```python
fig = plt.figure(figsize=[15,6])
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# def animate():
ax1.plot(eat2)
ax1.set_title("eat_count")
ax2.plot(coll1)
ax2.set_title("collision_count")
fig.savefig("Plot/plot.svg", dpi=150)
fig.savefig("Plot/plot.png", dpi=150)
plt.show()
# animate()
```
